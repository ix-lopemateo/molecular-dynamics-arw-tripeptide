```python

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("dist_ALA_TRP_298.xvg")

time = data[:, 0]
dist = data[:, 1]

plt.figure()
plt.plot(time, dist)
plt.xlabel("Time (ps)")
plt.ylabel("Distance (nm)")
plt.title("End-to-end distance (ALA-TRP, 298 K)")
plt.savefig("dist_ALA_TRP_298.png", dpi=300)
plt.show()
```

```python
data = np.loadtxt("dist_ALA_TRP_400.xvg")

time = data[:, 0]
dist = data[:, 1]

plt.figure()
plt.plot(time, dist, color='#E75480')  

plt.xlabel("Time (ps)")
plt.ylabel("Distance (nm)")
plt.title("End-to-end distance (ALA-TRP, 400 K)")

plt.savefig("dist_ALA_TRP_400.png", dpi=300)
plt.show()
```

COMPARACIÓN 

```python

# 298K
data_298 = np.loadtxt("dist_ALA_TRP_298.xvg")
time_298 = data_298[:, 0]
dist_298 = data_298[:, 1]

# 400K
data_400 = np.loadtxt("dist_ALA_TRP_400.xvg")
time_400 = data_400[:, 0]
dist_400 = data_400[:, 1]

# Figura única
plt.figure()

# Curvas
plt.plot(time_298, dist_298, label='298 K')
plt.plot(time_400, dist_400, label='400 K', color='#E75480')

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Distance (nm)")
plt.title("End-to-end distance comparison (298 K vs 400 K)")

# Leyenda
plt.legend()

# Opcional (queda mejor)
plt.grid(alpha=0.3)

# Guardar
plt.savefig("dist_ALA_TRP_comparison.png", dpi=300)

# Mostrar
plt.show()
```

```python
# 298K
data_298 = np.loadtxt("dist_ALA_ARG_298.xvg")
time_298 = data_298[:, 0]
dist_298 = data_298[:, 1]

# 400K
data_400 = np.loadtxt("dist_ALA_ARG_400.xvg")
time_400 = data_400[:, 0]
dist_400 = data_400[:, 1]

# Figura única
plt.figure()

# Curvas
plt.plot(time_298, dist_298, label='298 K')
plt.plot(time_400, dist_400, label='400 K', color='#E75480')

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Distance (nm)")
plt.title("ALA–ARG distance comparison (298 K vs 400 K))")

# Leyenda
plt.legend()

# Opcional (queda mejor)
plt.grid(alpha=0.3)

# Guardar
plt.savefig("dist_ALA_ARG_comparison.png", dpi=300)

# Mostrar
plt.show()
```
