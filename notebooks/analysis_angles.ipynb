```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("angle_298.xvg", comments=["@", "#"])

time = data[:, 0]
angle = data[:, 1]

plt.figure()
plt.plot(time, angle)
plt.xlabel("Time (ps)")
plt.ylabel("Angle (degrees)")
plt.title("Backbone angle (ALA–ARG–TRP), 298k")
plt.savefig("angle_backbone_298.png", dpi=300)
plt.show()
```

```python
data = np.loadtxt("angle_backbone_400.xvg", comments=["@", "#"])

time = data[:, 0]
angle = data[:, 1]

plt.figure()
plt.plot(time, angle, color='#E75480')

plt.xlabel("Time (ps)")
plt.ylabel("Angle (degrees)")
plt.title("Backbone angle (ALA–ARG–TRP), 400 K")

plt.savefig("angle_backbone_400.png", dpi=300)
plt.show()
```

```python
data = np.loadtxt("angle_ARG_TRP_CT3_400.xvg", comments=["@", "#"])

time = data[:, 0]
angle = data[:, 1]

plt.figure()
plt.plot(time, angle, color='#E75480')
plt.xlabel("Time (ps)")
plt.ylabel("Angle (degrees)")
plt.title("ARG–TRP–CT3 angle (400k)")
plt.savefig("angle_ARG_TRP_CT3_400.png", dpi=300)
plt.show()
```

```python
# 298K
data_298 = np.loadtxt("angle_298.xvg", comments=["@", "#"])
time_298 = data_298[:, 0]
angle_298 = data_298[:, 1]

# 400K
data_400 = np.loadtxt("angle_backbone_400.xvg", comments=["@", "#"])
time_400 = data_400[:, 0]
angle_400 = data_400[:, 1]

# Figura única
plt.figure()

# Curvas
plt.plot(time_298, angle_298, label='298 K')
plt.plot(time_400, angle_400, label='400 K', color='#E75480')

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Angle (degrees)")
plt.title("Backbone angle comparison (ALA–ARG–TRP), 298 K vs 400 K")

# Leyenda
plt.legend()

# Opcional (queda más limpio)
plt.grid(alpha=0.3)

# Guardar
plt.savefig("angle_backbone_comparison.png", dpi=300)

# Mostrar
plt.show()
```
