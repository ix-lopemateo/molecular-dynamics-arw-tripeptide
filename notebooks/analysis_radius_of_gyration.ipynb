RADIO DE GIRO 


Simulacion a 400k

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos
data = np.loadtxt("gyrate_400.xvg")

time = data[:, 0]
rg = data[:, 1]

# Crear figura
plt.figure()
plt.plot(time, rg, color='#E75480')

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Radius of gyration (nm)")
plt.title("Radius of gyration vs time (400 K)")

# Guardar figura
plt.savefig("gyrate_400.png", dpi=300)

# Mostrar
plt.show()
```

Simulacion a 298k

```python
# Cargar datos
data = np.loadtxt("gyrate.xvg")

time = data[:, 0]
rg = data[:, 1]

# Crear figura
plt.figure()
plt.plot(time, rg)

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Radius of gyration (nm)")
plt.title("Radius of gyration vs time (298 K)")

# Guardar figura
plt.savefig("gyrate_298.png", dpi=300)

# Mostrar
plt.show()
```

Comparación

```python

# Cargar 400K
data_400 = np.loadtxt("gyrate_400.xvg")
time_400 = data_400[:, 0]
rg_400 = data_400[:, 1]

# Cargar 298K
data_298 = np.loadtxt("gyrate.xvg")
time_298 = data_298[:, 0]
rg_298 = data_298[:, 1]

# Crear figura única
plt.figure()

# Dibujar ambas curvas
plt.plot(time_298, rg_298, label='298 K')
plt.plot(time_400, rg_400, label='400 K', color='#E75480')

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Radius of gyration (nm)")
plt.title("Radius of gyration vs time (298 K vs 400 K)")

# Leyenda (clave)
plt.legend()

# Guardar
plt.savefig("gyrate_comparison.png", dpi=300)

# Mostrar
plt.show()
```
