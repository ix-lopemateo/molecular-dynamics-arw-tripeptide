SIMULACION A 298K


```python
import numpy as np
import matplotlib.pyplot as plt

#TOTAL ENERGY
data = np.loadtxt("total_energy_298.xvg")

time = data[:, 0]
energy = data[:, 1]

plt.plot(time, energy)
plt.xlabel("Time (ps)")
plt.ylabel("Total Energy (kJ/mol)")
plt.title("Total Energy vs time (298 K)")
plt.savefig("total_energy_298.png", dpi=300)
plt.show()
```

```python
#KINETIC ENERGY
data = np.loadtxt("kinetic_energy_298.xvg")

time = data[:, 0]
ekin = data[:, 1]

plt.plot(time, ekin)
plt.xlabel("Time (ps)")
plt.ylabel("Kinetic Energy (kJ/mol)")
plt.title("Kinetic Energy vs time (298 K)")
plt.savefig("kinetic_energy_298.png", dpi=300)
plt.show()
```

SIMULACION A 400K

```python
#TOTAL ENERGY
data = np.loadtxt("total_energy_400.xvg")

time = data[:, 0]
energy = data[:, 1]

plt.plot(time, energy, color='#E75480')
plt.xlabel("Time (ps)")
plt.ylabel("Total Energy (kJ/mol)")
plt.title("Total Energy vs time (400 K)")
plt.savefig("total_energy_400.png", dpi=300)
plt.show()
```

```python
#KINETIC ENERGY
data = np.loadtxt("kinetic_energy_400.xvg")

time = data[:, 0]
energy = data[:, 1]

plt.plot(time, energy, color='#E75480')
plt.xlabel("Time (ps)")
plt.ylabel("Kinetic Energy (kJ/mol)")
plt.title("Kinetic Energy vs time (400 K)")
plt.savefig("kinetic_energy_400.png", dpi=300)
plt.show()
```

COMPARACIONES

```python
# 298K
data_298 = np.loadtxt("total_energy_298.xvg")
time_298 = data_298[:, 0]
energy_298 = data_298[:, 1]

# 400K
data_400 = np.loadtxt("total_energy_400.xvg")
time_400 = data_400[:, 0]
energy_400 = data_400[:, 1]

# Figura única
plt.figure()

# Curvas
plt.plot(time_298, energy_298, label='298 K')
plt.plot(time_400, energy_400, label='400 K', color='#E75480')

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Total Energy (kJ/mol)")
plt.title("Total Energy vs time")

# Leyenda
plt.legend()

# Opcional
plt.grid(alpha=0.3)

# Guardar
plt.savefig("total_energy_comparison.png", dpi=300)

# Mostrar
plt.show()
```

```python
# 298K
data_298 = np.loadtxt("kinetic_energy_298.xvg")
time_298 = data_298[:, 0]
ekin_298 = data_298[:, 1]

# 400K
data_400 = np.loadtxt("kinetic_energy_400.xvg")
time_400 = data_400[:, 0]
ekin_400 = data_400[:, 1]

# Figura única
plt.figure()

# Curvas
plt.plot(time_298, ekin_298, label='298 K')
plt.plot(time_400, ekin_400, label='400 K', color='#E75480')

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Kinetic Energy (kJ/mol)")
plt.title("Kinetic Energy vs time")

# Leyenda
plt.legend()

# Opcional
plt.grid(alpha=0.3)

# Guardar
plt.savefig("kinetic_energy_comparison.png", dpi=300)

# Mostrar
plt.show()
```
