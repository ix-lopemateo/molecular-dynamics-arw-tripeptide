```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("temperature_298.xvg")

temp = data[:, 1]

plt.hist(temp, bins=30, edgecolor='black')
plt.xlabel("Temperature (K)")
plt.ylabel("Frequency")
plt.title("Temperature distribution (298 K)")
plt.savefig("temperature_hist_298.png", dpi=300)
plt.show()
```

```python
data = np.loadtxt("temperature_400.xvg")

temp = data[:, 1]

plt.hist(temp, bins=30, edgecolor='black', color='#E75480')
plt.xlabel("Temperature (K)")
plt.ylabel("Frequency")
plt.title("Temperature distribution (400 K)")

plt.savefig("temperature_hist_400.png", dpi=300)
plt.show()
```

```python
# 298K
data_298 = np.loadtxt("temperature_298.xvg")
temp_298 = data_298[:, 1]

# 400K
data_400 = np.loadtxt("temperature_400.xvg")
temp_400 = data_400[:, 1]

plt.figure()

# Histogramas superpuestos
plt.hist(temp_298, bins=30, edgecolor='black', label='298 K')
plt.hist(temp_400, bins=30, edgecolor='black', color='#E75480', label='400 K')

# Etiquetas
plt.xlabel("Temperature (K)")
plt.ylabel("Frequency")
plt.title("Temperature distribution comparison, 298 K vs 400 K")

# Leyenda
plt.legend()

# Guardar
plt.savefig("temperature_comparison.png", dpi=300)

# Mostrar
plt.show()
```
