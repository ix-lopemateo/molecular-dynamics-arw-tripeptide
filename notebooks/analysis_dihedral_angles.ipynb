```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("dihedral_PHI_298.xvg", comments=["@", "#"])

time = data[:, 0]
angle = data[:, 1]

plt.plot(time, angle)
plt.xlabel("Time (ps)")
plt.ylabel("Dihedral angle (degrees)")
plt.title("Dihedral angle PHI (ARG)")
plt.savefig("dihedral_PHI_ARG_298.png", dpi=300)
plt.show()
```

```python
data = np.loadtxt("dihedral_PSI_298.xvg", comments=["@", "#"])

time = data[:, 0]
angle = data[:, 1]

plt.plot(time, angle)
plt.xlabel("Time (ps)")
plt.ylabel("Dihedral angle (degrees)")
plt.title("Dihedral angle PSI (ARG)")
plt.savefig("dihedral_PSI_ARG_298.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# PHI
data_phi = np.loadtxt("dihedral_PHI_298.xvg", comments=["@", "#"])
time_phi = data_phi[:, 0]
phi = data_phi[:, 1]

# PSI
data_psi = np.loadtxt("dihedral_PSI_298.xvg", comments=["@", "#"])
time_psi = data_psi[:, 0]
psi = data_psi[:, 1]

# Figura única
plt.figure()

# Curvas
plt.plot(time_phi, phi, label='PHI (ARG)')
plt.plot(time_psi, psi, label='PSI (ARG)', color='#AEC6CF')  # azul pastel

# Etiquetas
plt.xlabel("Time (ps)")
plt.ylabel("Dihedral angle (degrees)")
plt.title("Dihedral angles (ARG, 298 K)")

# Leyenda
plt.legend()

# Opcional
plt.grid(alpha=0.3)

# Guardar
plt.savefig("dihedral_ARG_298.png", dpi=300)

# Mostrar
plt.show()
```
