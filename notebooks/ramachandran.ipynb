```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm 

phi = []
psi = []

with open("rama_opt.xvg") as f:
    for line in f:
        parts = line.split()
        if len(parts) < 3:
            continue
        try:
            phi.append(float(parts[0]))
            psi.append(float(parts[1]))
        except:
            continue

phi = np.array(phi)
psi = np.array(psi)

plt.figure(figsize=(6,5))

# histograma 2D 
hist = plt.hist2d(phi, psi, bins=60)

plt.xlabel("φ (grados)")
plt.ylabel("ψ (grados)")
plt.title("Diagrama Ramachandran 2D (298 K)")

# barra de color
plt.colorbar(label="Frecuencia")

plt.xlim(-180, 180)
plt.ylim(-180, 180)

plt.savefig("rama2D_298_opt2.png", dpi=300)
plt.show()
```
