```python
import numpy as np
import matplotlib.pyplot as plt

files = [
    ("vel_ALA_CA_298.xvg", "ALA CA"),
    ("vel_ARG_N_298.xvg", "ARG N"),
    ("vel_ARG_CA_298.xvg", "ARG CA"),
    ("vel_ARG_C_298.xvg", "ARG C"),
    ("vel_TRP_CA_298.xvg", "TRP CA"),
]

for f, label in files:
    data = np.loadtxt(f, comments=["@", "#"])
    vel = data[:,1]
    plt.hist(vel, bins=50, alpha=0.5, label=label)

plt.xlabel("Velocity (nm/ps)")
plt.ylabel("Frequency")
plt.legend()
plt.title("Velocity distributions (298 K)")
plt.show()
```
